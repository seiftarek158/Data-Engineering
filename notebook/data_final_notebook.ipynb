## Filling Null Values

````python
# First we will convert the timestamp to date in order to compare it with the dtp
trades['date'] = pd.to_datetime(trades['timestamp']).dt.date

#Then we will calculate the estimated prices per date per stock from the trades dataframe
estimated_prices_per_date_per_stock = trades.groupby(['date', 'stock_ticker']).apply(
    lambda x: (x['cumulative_portfolio_value'] / x['quantity'] * x['quantity']).sum() / x['quantity'].sum()
).reset_index(name='price')
len(estimated_prices_per_date_per_stock)
````

````python
# Fill missing values for each stock column in the dtp 
for col in dtp.columns:
    if col == 'date':
        continue
    
    # Get trade prices for this stock
    stock_prices = estimated_prices_per_date_per_stock[estimated_prices_per_date_per_stock['stock_ticker'] == col].set_index('date')['price']
    
    # Fill missing values from trades by checking the null rows and then mapping it to the correct stockprice using the date
    missing_mask = dtp[col].isnull()
    dtp.loc[missing_mask, col] = dtp.loc[missing_mask, 'date'].map(stock_prices)
    
    # Fill any remaining with forward fill
    dtp[col] = dtp[col].fillna(method='ffill')
````

## Removing Outliers

````python
# First we will preprocess the dtp dataset
dtp['date'] = pd.to_datetime(dtp['date'])
rows = []

# what we are doing here is converting so instead of having one row per date with multiple stock columns, we will have multiple rows per date, each with a single stock ticker and its price
for _, row in dtp.iterrows():
    date = row['date']
    for ticker in dtp.columns:
        if ticker != 'date':
            rows.append({
                'date': date,
                'stock_ticker': ticker,
                'stock_price': row[ticker]
            })

dtp_long = pd.DataFrame(rows)
# Check dtp_long
print(f"DTP long format shape: {dtp_long.shape}")
dtp_long.head()
````

````python
cols = ['quantity', 'average_trade_size', 'cumulative_portfolio_value']

for col in cols:
    s = trades[col]
    sk = s.skew()
    print(f'Column: {col} | skewness={sk:.3f}')
    # interpret skewness magnitude
    if abs(sk) > 1:
        interp = 'highly skewed'
    elif abs(sk) > 0.5:
        interp = 'moderately skewed'
    else:
        interp = 'approximately symmetric'
    print('->', interp)
````

````python
# Per-column boxplots + Z-score outlier counts and percentages
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

col = 'quantity'

series = trades[col]
n = len(series)


# Calculate IQR-based outliers
q1 = series.quantile(0.25)
q3 = series.quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
outliers_mask = (series < lower) | (series > upper)
n_outliers = outliers_mask.sum()
pct_outliers = (n_outliers / n) * 100

print(f'Column: {col}')
print(f'Non-null count: {n}')
print(f'Outliers (IQR method): {n_outliers} ({pct_outliers:.2f}%)')
print(f'Q1: {q1:.2f}, Q3: {q3:.2f}')
print(f'Lower fence: {lower:.2f}, Upper fence: {upper:.2f}')

# Create boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(y=series, color='skyblue')
plt.title(f'{col} - Outliers: {n_outliers} ({pct_outliers:.2f}%)')
plt.ylabel('Value')

# Add horizontal lines for fences
plt.axhline(lower, color='red', linestyle='--', linewidth=0.8, label='lower fence')
plt.axhline(upper, color='red', linestyle='--', linewidth=0.8, label='upper fence')
plt.axhline(q1, color='green', linestyle=':', linewidth=0.7, label='Q1')
plt.axhline(q3, color='green', linestyle=':', linewidth=0.7, label='Q3')
plt.legend(loc='upper right', fontsize=8)

plt.tight_layout()
plt.show()
````

## Data Inconsistencies Check

````python
#Check for Inconsistencies
print("=== INCONSISTENCIES ===")
total_issues = 0

# Check if any of the tables of the duplicated
dup_trans_id = trades['transaction_id'].duplicated().sum()
dup_cust_id = dc['customer_id'].duplicated().sum()
dup_stock_ticker = ds['stock_ticker'].duplicated().sum()
dup_date = dd['date'].duplicated().sum()
# Check for missing values
# Check if any customer_id in trades not in dim_customer table (referential integrity)
missing_cust = len(trades[~trades['customer_id'].isin(dc['customer_id'])]['customer_id'].unique())
# Check if any stock in trades not in stock ticker
missing_stock = len(trades[~trades['stock_ticker'].isin(ds['stock_ticker'])]['stock_ticker'].unique())
# Check if any date of a trade not in dim_date table
trades_temp = trades.copy()
trades_temp['date'] = pd.to_datetime(trades_temp['timestamp']).dt.date
dd_temp = dd.copy()
dd_temp['date'] = pd.to_datetime(dd_temp['date']).dt.date
missing_date = len(trades_temp[~trades_temp['date'].isin(dd_temp['date'])]['date'].unique())
# check for invalid like negative quantities
neg_qty = (trades['quantity'] < 0).sum()
neg_avg_trade = (trades['average_trade_size'] < 0).sum()
neg_portfolio = (trades['cumulative_portfolio_value'] < 0).sum()

total_issues = dup_trans_id + dup_cust_id + dup_stock_ticker + dup_date + missing_cust + missing_stock + missing_date + neg_qty + neg_avg_trade + neg_portfolio

print(f"Total inconsistencies found: {total_issues}")
print(f"- Duplicate transaction IDs: {dup_trans_id}")
print(f"- Duplicate customer IDs: {dup_cust_id}")
print(f"- Duplicate stock tickers: {dup_stock_ticker}")
print(f"- Duplicate dates: {dup_date}")
print(f"- Missing customer references: {missing_cust}")
print(f"- Missing stock references: {missing_stock}")
print(f"- Missing date references: {missing_date}")
print(f"- Negative quantities: {neg_qty}")
print(f"- Negative average trade sizes: {neg_avg_trade}")
print(f"- Negative portfolio values: {neg_portfolio}")
````

````python
# Now check that the date column is consistent with the actual date

print("\n=== DIM_DATE CONSISTENCY CHECK ===")

dd_check = dd.copy()
dd_check['date'] = pd.to_datetime(dd_check['date'])

# Extract actual day name and weekend status from date
dd_check['actual_day_name'] = dd_check['date'].dt.day_name()
dd_check['actual_is_weekend'] = dd_check['date'].dt.dayofweek.isin([5, 6]).astype(int)

# Check mismatches
day_name_mismatch = (dd_check['day_name'] != dd_check['actual_day_name']).sum()
is_weekend_mismatch = (dd_check['is_weekend'] != dd_check['actual_is_weekend']).sum()

print(f"Day name mismatches: {day_name_mismatch}")
print(f"Is_weekend mismatches: {is_weekend_mismatch}")

if day_name_mismatch > 0:
    print("\nSample day_name mismatches:")
    print(dd_check[dd_check['day_name'] != dd_check['actual_day_name']][['date', 'day_name', 'actual_day_name']].head())

if is_weekend_mismatch > 0:
    print("\nSample is_weekend mismatches:")
    print(dd_check[dd_check['is_weekend'] != dd_check['actual_is_weekend']][['date', 'day_name', 'is_weekend', 'actual_is_weekend']].head())

print(f"\nTotal date-related inconsistencies: {day_name_mismatch + is_weekend_mismatch}")
````

## Part 1: Encoding

### 1.1 Data Encoding
Encode the following columns:
1. **stock_ticker** - Label Encoding (20 unique stock tickers)
2. **transaction_type** - binary Encoding (buy/sell â†’ 0/1)
3. **customer_account_type** - binary Encoding (Institutional/Retail â†’ 0/1)
4. **day_name** - One-Hot Encoding (7 days of the week - nominal categorical)
5. **is_weekend** - Boolean to Binary (True/False â†’ 1/0)
6. **is_holiday** - Boolean to Binary (True/False â†’ 1/0)
7. **stock_sector** - Label Encoding (multiple sectors - categorical)
8. **stock_industry** - Label Encoding (multiple industries - categorical)

````python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

def encode_data(df):
    """
    Encodes categorical columns in the integrated dataframe.
    
    Label Encoding (modifies original columns):
    - stock_ticker, transaction_type, customer_account_type, stock_sector
    
    One-Hot Encoding (creates new columns, removes original):
    - day_name, stock_industry
    
    Boolean to Binary (modifies original columns):
    - is_weekend, is_holiday
    
    Parameters:
    -----------
    df : pandas.DataFrame
        The integrated dataframe from milestone 1
    
    Returns:
    --------
    pandas.DataFrame
        The dataframe with encoded columns
    """
    # Create a copy to avoid modifying the original dataframe
    df_encoded = df.copy()
    
    # Initialize label encoders
    le_stock = LabelEncoder()
    le_transaction = LabelEncoder() # act as a binary encoder here
    le_account = LabelEncoder() # act as a binary encoder here
    le_sector = LabelEncoder()
    le_industry = LabelEncoder()
    
   # Apply Label Encoding - MODIFY ORIGINAL COLUMNS
    # 1. stock_ticker - 20 unique values (STK001-STK020)
    df_encoded['stock_ticker'] = le_stock.fit_transform(df_encoded['stock_ticker'])
    
    # 2. stock_sector - Multiple sectors
    df_encoded['stock_sector'] = le_sector.fit_transform(df_encoded['stock_sector'])

    df_encoded['stock_industry'] = le_industry.fit_transform(df_encoded['stock_industry'])
    
    # Apply One-Hot Encoding
    # 3. day_name - One-Hot Encoding (nominal categorical with 7 categories)
    day_dummies = pd.get_dummies(df_encoded['day_name'], prefix='day', dtype=int)
    df_encoded = pd.concat([df_encoded, day_dummies], axis=1)
    df_encoded.drop('day_name', axis=1, inplace=True)  # Remove original column
    
    
    
    # Boolean to Binary - MODIFY ORIGINAL COLUMNS
    # 4. is_weekend - Convert boolean to binary (True/False â†’ 1/0)
    df_encoded['is_weekend'] = df_encoded['is_weekend'].astype(int)
    
    # 5. is_holiday - Convert boolean to binary (True/False â†’ 1/0)
    df_encoded['is_holiday'] = df_encoded['is_holiday'].astype(int)
    
    # 6. transaction_type - 2 unique values (buy/sell)
    df_encoded['transaction_type'] = le_transaction.fit_transform(df_encoded['transaction_type'])
    
    # 7. customer_account_type - 2 unique values (Institutional/Retail)
    df_encoded['customer_account_type'] = le_account.fit_transform(df_encoded['customer_account_type'])
    
    # Print encoding information
    print("=" * 70)
    print("ENCODING SUMMARY")
    print("=" * 70)
    
    print(f"\n1. stock_ticker (Label Encoding - Original Column Modified):")
    print(f"   - Unique values encoded: {len(le_stock.classes_)}")
    print(f"   - Encoded range: 0-{len(le_stock.classes_)-1}")
    print(f"   - Sample mapping: {dict(list(zip(le_stock.classes_[:5], range(5))))}")
    
    print(f"\n2. transaction_type (Label Encoding - Original Column Modified):")
    print(f"   - Unique values: {len(le_transaction.classes_)}")
    print(f"   - Mapping: {dict(zip(le_transaction.classes_, le_transaction.transform(le_transaction.classes_)))}")
    
    print(f"\n3. customer_account_type (Label Encoding - Original Column Modified):")
    print(f"   - Unique values: {len(le_account.classes_)}")
    print(f"   - Mapping: {dict(zip(le_account.classes_, le_account.transform(le_account.classes_)))}")
    
    print(f"\n4. stock_sector (Label Encoding - Original Column Modified):")
    print(f"   - Unique sectors: {len(le_sector.classes_)}")
    print(f"   - Sectors: {sorted(le_sector.classes_)}")
    print(f"   - Encoded range: 0-{len(le_sector.classes_)-1}")
    
    print(f"\n5. day_name (One-Hot Encoding - Original column removed):")
    print(f"   - Unique values: {len(day_dummies.columns)}")
    print(f"   - New columns created: {list(day_dummies.columns)}")
    print(f"   - Values: 0 or 1 (binary)")
    
    print(f"\n6. stock_industry (One-Hot Encoding - Original column removed):")
    print(f"   - Unique values: {len(industry_dummies.columns)}")
    print(f"   - New columns created: {list(industry_dummies.columns)}")
    print(f"   - Values: 0 or 1 (binary)")
    
    print(f"\n7. is_weekend (Boolean â†’ Binary - Original Column Modified):")
    print(f"   - Converted to: Trueâ†’1, Falseâ†’0")
    print(f"   - Unique values: {sorted(df_encoded['is_weekend'].unique())}")
    
    print(f"\n8. is_holiday (Boolean â†’ Binary - Original Column Modified):")
    print(f"   - Converted to: Trueâ†’1, Falseâ†’0")
    print(f"   - Unique values: {sorted(df_encoded['is_holiday'].unique())}")
    
    print(f"\n{'=' * 70}")
    print(f"Original columns: {len(df.columns)}")
    print(f"Encoded columns: {len(df_encoded.columns)}")
    print(f"Columns removed: 1 (day_name)")
    print("=" * 70)
    
    # Store the encoders as attributes for potential inverse transformation
    df_encoded.attrs['encoders'] = {
        'stock_ticker': le_stock,
        'transaction_type': le_transaction,
        'customer_account_type': le_account,
        'stock_sector': le_sector,
        'stock_industry': le_industry
    }
    
    return df_encoded
````

````python
def create_encoding_lookup_tables(df_original, df_encoded):
    """
    Creates lookup tables for all encoded columns showing the mapping
    between original and encoded values.
    
    Since label encoding now modifies original columns, we use the encoders
    stored in df_encoded.attrs['encoders'] to retrieve the mappings.
    
    Parameters:
    -----------
    df_original : pandas.DataFrame
        The original dataframe before encoding
    df_encoded : pandas.DataFrame
        The encoded dataframe with encoded columns
    
    Returns:
    --------
    dict
        Dictionary of lookup tables for each encoded column
    """
    lookup_tables = {}
    
    # Get the encoders from the encoded dataframe attributes
    encoders = df_encoded.attrs.get('encoders', {})
    
    # 1. Stock Ticker Lookup (Label Encoding)
    if 'stock_ticker' in encoders:
        le_stock = encoders['stock_ticker']
        stock_lookup = pd.DataFrame({
            'Column Name': 'stock_ticker',
            'Original Value': le_stock.classes_,
            'Encoded Value': le_stock.transform(le_stock.classes_)
        })
        lookup_tables['stock_ticker'] = stock_lookup.sort_values('Encoded Value')
    
    # 2. Transaction Type Lookup (Label Encoding)
    if 'transaction_type' in encoders:
        le_trans = encoders['transaction_type']
        trans_lookup = pd.DataFrame({
            'Column Name': 'transaction_type',
            'Original Value': le_trans.classes_,
            'Encoded Value': le_trans.transform(le_trans.classes_)
        })
        lookup_tables['transaction_type'] = trans_lookup.sort_values('Encoded Value')
    
    # 3. Customer Account Type Lookup (Label Encoding)
    if 'customer_account_type' in encoders:
        le_account = encoders['customer_account_type']
        account_lookup = pd.DataFrame({
            'Column Name': 'customer_account_type',
            'Original Value': le_account.classes_,
            'Encoded Value': le_account.transform(le_account.classes_)
        })
        lookup_tables['customer_account_type'] = account_lookup.sort_values('Encoded Value')
    # 4. Stock Industry Lookup (Label Encoding)
    if 'stock_industry' in encoders:
        le_industry = encoders['stock_industry']
        industry_lookup = pd.DataFrame({
            'Column Name': 'stock_industry',
            'Original Value': le_industry.classes_,
            'Encoded Value': le_industry.transform(le_industry.classes_)
        })
        lookup_tables['stock_industry'] = industry_lookup.sort_values('Encoded Value')
    # 5. Stock Sector Lookup (Label Encoding)
    if 'stock_sector' in encoders:
        le_sector = encoders['stock_sector']
        sector_lookup = pd.DataFrame({
            'Column Name': 'stock_sector',
            'Original Value': le_sector.classes_,
            'Encoded Value': le_sector.transform(le_sector.classes_)
        })
        lookup_tables['stock_sector'] = sector_lookup.sort_values('Encoded Value')
    
    # 6. Day Name Lookup (One-Hot Encoding)
    day_cols = [col for col in df_encoded.columns if col.startswith('day_')]
    day_lookup_data = []
    for day_col in sorted(day_cols):
        day_name = day_col.replace('day_', '')
        day_lookup_data.append({
            'Column Name': 'day_name',
            'Original Value': day_name,
            'Encoded Column': day_col
        })
    if day_lookup_data:
        day_lookup = pd.DataFrame(day_lookup_data)
        lookup_tables['day_name'] = day_lookup
    
    # 7. Is Weekend Lookup (Boolean to Binary)
    weekend_lookup = pd.DataFrame({
        'Column Name': ['is_weekend', 'is_weekend'],
        'Original Value': [False, True],
        'Encoded Value': [0, 1]
    })
    lookup_tables['is_weekend'] = weekend_lookup
    
    # 8. Is Holiday Lookup (Boolean to Binary)
    holiday_lookup = pd.DataFrame({
        'Column Name': ['is_holiday', 'is_holiday'],
        'Original Value': [False, True],
        'Encoded Value': [0, 1]
    })
    lookup_tables['is_holiday'] = holiday_lookup
    
    return lookup_tables

def create_imputation_lookup_table(dtp_original):
    """
    Creates a lookup table for imputed values showing original and imputed values.
    
    Parameters:
    -----------
    dtp_original : pandas.DataFrame
        Original daily trade prices dataframe before imputation (wide format)
    
    Returns:
    --------
    pandas.DataFrame
        Lookup table showing which values were imputed
    """
    imputation_records = []
    
    # Check DTP imputation (stock prices) - working with wide format
    for col in dtp_original.columns:
        if col == 'date':
            continue
        
        # Find rows that had null values
        null_mask = dtp_original[col].isnull()
        null_count = null_mask.sum()
        
        if null_count > 0:
            # Get the dates where imputation occurred
            null_dates = dtp_original.loc[null_mask, 'date'].values
            
            # Add a summary record for this stock
            imputation_records.append({
                'Column Name': f'stock_price ({col})',
                'Stock Ticker': col,
                'Number of Nulls': null_count,
                'Imputation Method': 'Trade-based calculation â†’ Forward fill',
                'Sample Dates': str(null_dates[:5]) if len(null_dates) > 5 else str(null_dates)
            })
    
    imputation_lookup = pd.DataFrame(imputation_records)
    return imputation_lookup

def save_lookup_tables(lookup_tables, imputation_lookup, save_format='csv'):
    """
    Saves all lookup tables to files or database.
    
    Parameters:
    -----------
    lookup_tables : dict
        Dictionary of encoding lookup tables
    imputation_lookup : pandas.DataFrame
        Imputation lookup table
    save_format : str
        'csv' or 'db' for database
    """
    import os
    
    # Create lookups directory if it doesn't exist
    os.makedirs('data/lookups', exist_ok=True)
    
    if save_format == 'csv':
        # Save encoding lookup tables
        for name, table in lookup_tables.items():
            filename = f'data/lookups/encoding_lookup_{name}.csv'
            table.to_csv(filename, index=False)
            print(f"Saved: {filename}")
        
        # Save imputation lookup table
        if not imputation_lookup.empty:
            imputation_lookup.to_csv('data/lookups/imputation_lookup.csv', index=False)
            print(f"Saved: data/lookups/imputation_lookup.csv")
        
        # Collect all encoding lookup tables: label-encoded, one-hot and binary
        label_keys = ['stock_ticker', 'stock_sector','stock_industry']
        onehot_keys = ['day_name']
        binary_keys = ['is_weekend', 'is_holiday','transaction_type', 'customer_account_type',]
        selected_keys = label_keys + onehot_keys + binary_keys

        # Keep only keys that exist in lookup_tables
        label_encoded_tables = [v for k, v in lookup_tables.items() if k in selected_keys]
        if label_encoded_tables:
            master_lookup = pd.concat(label_encoded_tables, ignore_index=True)
            master_lookup.to_csv('data/lookups/master_encoding_lookup.csv', index=False)
            print(f"Saved: data/lookups/master_encoding_lookup.csv")
        
    elif save_format == 'db':
        from db_utils import save_to_db
        
        # Save each lookup table to database
        for name, table in lookup_tables.items():
            save_to_db(table, f'lookup_encoding_{name}')
            print(f"Saved to DB: lookup_encoding_{name}")
        
        # Save imputation lookup
        if not imputation_lookup.empty:
            save_to_db(imputation_lookup, 'lookup_imputation')
            print(f"Saved to DB: lookup_imputation")
        
        # Save master lookup (only label-encoded columns)
        label_encoded_tables = [v for k, v in lookup_tables.items() 
                                if k in ['stock_ticker', 'transaction_type', 'customer_account_type', 'stock_sector',
                                         'stock_industry']]
        if label_encoded_tables:
            master_lookup = pd.concat(label_encoded_tables, ignore_index=True)
            save_to_db(master_lookup, 'lookup_master_encoding')
            print(f"Saved to DB: lookup_master_encoding")

# Create lookup tables
print("Creating encoding lookup tables...")
encoding_lookups = create_encoding_lookup_tables(integrated, integrated_encoded)

# Display each lookup table
for name, table in encoding_lookups.items():
    print(f"\n{'='*70}")
    print(f"LOOKUP TABLE: {name.upper()}")
    print('='*70)
    print(table.to_string(index=False))

# Create imputation lookup table
print("\n\nCreating imputation lookup table...")
# Load the original DTP before imputation
dtp_original = pd.read_csv('data/daily_trade_prices.csv')

imputation_lookup = create_imputation_lookup_table(dtp_original)

print(f"\n{'='*70}")
print("IMPUTATION LOOKUP TABLE")
print('='*70)
if not imputation_lookup.empty:
    print(imputation_lookup.to_string(index=False))
save_lookup_tables(encoding_lookups, imputation_lookup, save_format='csv')
````

## Kafka Streaming Consumer

This section implements a Kafka consumer function that processes streamed data row-by-row from the `stream.csv` file. The consumer:
1. Subscribes to the Kafka topic
2. Receives records one at a time
3. Preprocesses each record (encoding categorical variables)
4. Stops when EOS (End of Stream) message is received
5. Combines the streamed data with the main dataset

````python
import json
from kafka import KafkaConsumer

def encode_row(row, encoding_lookups):
    """
    Encodes a single row/record with the same logic as encode_data but for individual rows.
    
    Parameters:
    -----------
    row : dict
        A single record/row to encode
    encoding_lookups : dict
        Dictionary containing all encoding mappings:
        - 'stock_ticker': dict mapping ticker names to encoded values
        - 'transaction_type': dict mapping transaction types to encoded values
        - 'customer_account_type': dict mapping account types to encoded values
        - 'stock_sector': dict mapping sectors to encoded values
        - 'day_names': list of all possible day names
        - 'industry_names': list of all possible industry names
    
    Returns:
    --------
    dict
        The encoded row with all transformations applied
    """
    encoded_row = row.copy()
    
    # Label Encoding - Modify values directly using lookup dictionaries
    if 'stock_ticker' in encoded_row and encoded_row['stock_ticker'] is not None:
        encoded_row['stock_ticker'] = encoding_lookups['stock_ticker'].get(row['stock_ticker'], -1)
    
    if 'transaction_type' in encoded_row and encoded_row['transaction_type'] is not None:
        encoded_row['transaction_type'] = encoding_lookups['transaction_type'].get(row['transaction_type'], -1)
    
    if 'customer_account_type' in encoded_row and encoded_row['customer_account_type'] is not None:
        encoded_row['customer_account_type'] = encoding_lookups['customer_account_type'].get(row['customer_account_type'], -1)
    
    if 'stock_sector' in encoded_row and encoded_row['stock_sector'] is not None:
        encoded_row['stock_sector'] = encoding_lookups['stock_sector'].get(row['stock_sector'], -1)
    
    # One-Hot Encoding - Create new columns and remove original
    if 'day_name' in encoded_row and encoded_row['day_name'] is not None:
        day_value = encoded_row['day_name']
        for day in encoding_lookups['day_names']:
            encoded_row[f'day_{day}'] = 1 if day_value == day else 0
        del encoded_row['day_name']  # Remove original column
    
    if 'stock_industry' in encoded_row and encoded_row['stock_industry'] is not None:
        industry_value = encoded_row['stock_industry']
        for industry in encoding_lookups['industry_names']:
            encoded_row[f'industry_{industry}'] = 1 if industry_value == industry else 0
        del encoded_row['stock_industry']  # Remove original column
    
    # Boolean to Binary - Modify values directly
    if 'is_weekend' in encoded_row:
        encoded_row['is_weekend'] = int(encoded_row['is_weekend']) if encoded_row['is_weekend'] is not None else 0
    
    if 'is_holiday' in encoded_row:
        encoded_row['is_holiday'] = int(encoded_row['is_holiday']) if encoded_row['is_holiday'] is not None else 0
    
    return encoded_row

def consume_kafka_stream(topic_name='55_0654_Topic', bootstrap_servers=['localhost:9092'], 
                         main_data_path='data/integrated_main.csv',
                         unencoded_data_path='data/integrated_main_unencoded.csv', 
                         output_file='data/FULL_STOCKS.csv'):
    """
    Consumes Kafka stream data row-by-row, preprocessing each record until EOS is received.
    
    Parameters:
    -----------
    topic_name : str
        Kafka topic name to consume from
    bootstrap_servers : list
        List of Kafka bootstrap servers
    main_data_path : str
        Path to the encoded main dataset for final combination
    unencoded_data_path : str
        Path to the unencoded main dataset for reference or other uses
    output_file : str
        Path to save the final combined dataset
    
    Returns:
    --------
    pandas.DataFrame
        Combined dataset with streamed data appended
    """
    print("="*70)
    print("KAFKA CONSUMER - RECEIVING STREAMED DATA")
    print("="*70)
    print(f"Topic: {topic_name}")
    print("="*70 + "\n")

    
    try:
        # Load unencoded data to create encoding lookups
        print("Loading encoding lookups from unencoded main dataset...")
        unencoded_data = pd.read_csv(unencoded_data_path)
        
        encoding_lookups = {}
        
        # Create lookup dictionaries for label encoding
        encoding_lookups['stock_ticker'] = {val: idx for idx, val in enumerate(sorted(unencoded_data['stock_ticker'].unique()))}
        encoding_lookups['transaction_type'] = {val: idx for idx, val in enumerate(sorted(unencoded_data['transaction_type'].unique()))}
        encoding_lookups['customer_account_type'] = {val: idx for idx, val in enumerate(sorted(unencoded_data['customer_account_type'].unique()))}
        encoding_lookups['stock_sector'] = {val: idx for idx, val in enumerate(sorted(unencoded_data['stock_sector'].unique()))}
        
        # Create lists for one-hot encoding
        encoding_lookups['day_names'] = sorted(unencoded_data['day_name'].unique().tolist())
        encoding_lookups['industry_names'] = sorted(unencoded_data['stock_industry'].unique().tolist())
        
        print("âœ“ Encoding lookups created successfully")
        
        # Load encoded main data for final combination
        print("Loading encoded main dataset...")
        main_data = pd.read_csv(main_data_path)
        print(f"Main dataset: {len(main_data)} records")
        # Create consumer
        consumer = KafkaConsumer(
            topic_name,
            bootstrap_servers=bootstrap_servers,
            auto_offset_reset='latest',
            enable_auto_commit=True,
            group_id='stock-data-consumer',
            value_deserializer=lambda x: json.loads(x.decode('utf-8'))
        )

        print(f"âœ“ Consumer subscribed to topic: {topic_name}")
        print("Waiting for streamed data...")
        print("="*70)

        processed_records = []
        record_count = 0

        for message in consumer:
            record = message.value

            # Check for End of Stream
            if 'EOS' in record and record['EOS']:
                print("="*70)
                print("âœ“ Received EOS (End of Stream) message")
                break

            # Process the record row-by-row using encode_row function
            processed = encode_row(record, encoding_lookups)

            processed_records.append(processed)
            record_count += 1
            
            print(f"âœ“ Processed record {record_count}: Transaction ID {record.get('transaction_id')}")

        consumer.close()
        print("âœ“ Consumer closed")

        if processed_records:
            streamed_df = pd.DataFrame(processed_records)
            print(f"\nâœ“ Processed {len(streamed_df)} streamed records")
            
            full_dataset = pd.concat([main_data, streamed_df], ignore_index=True)
            print(f"âœ“ Combined dataset: {len(full_dataset)} total records")
            
            # Sort by transaction_id before saving
            print("\nâœ“ Sorting dataset by transaction_id...")
            full_dataset = full_dataset.sort_values(by='transaction_id').reset_index(drop=True)
            print(f"âœ“ Dataset sorted successfully")
            
            full_dataset.to_csv(output_file, index=False)
            print(f"âœ“ Saved to {output_file}")
            print(f"\nâœ“ Streaming processing completed!")
            print(f"Final dataset shape: {full_dataset.shape}")
            
            return full_dataset
        else:
            print("\nâš  No records received from stream")
            return main_data

    except Exception as e:
        print(f"\nâœ— Error occurred: {str(e)}")
        print("Make sure Kafka is running and producer has sent data")
        raise

print("âœ“ encode_row() and consume_kafka_stream() functions defined successfully")
````
